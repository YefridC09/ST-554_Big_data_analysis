First, we need to connect to the database and import the modules we will use.

```python

#Import the modules
import sqlite3
import pandas as pd

#Connectin to the .sqlite file
MLB = sqlite3.connect("lahman_1871-2022.sqlite")
```

I will create a function for query execution.

```python
def execute_query(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        print("Query executed successfully")
        #connection.commit()
    except sqlite3.Error as e:
        print(f"The error '{e}' occurred")
    cursor.close()
```

Now we look for the schema of the database using the function `read_sql()` from `pandas` module.

```python
get_schema = '''
        SELECT *
        FROM sqlite_schema
        WHERE type = "table";
        '''

# creating a dataframe with the schema
schema_MLB = pd.read_sql(get_schema, MLB)
schema_MLB
```

Now that we know which tables are in the database, we can start working with the data.

### 2

It is necessary to create a query that will do an inner join between the `HallOfFame` table and the `Pitching` table in the `playerID` variable, and return the `GS, G, W, L, IPOuts, CG, SHO` and `SV` columns.

```python
# Creating a query
Pit_HOF = """
    SELECT HallOfFame.playerID, GS, G, W, L, IPOuts, CG, SHO, SV
    From HallOfFame
    Inner JOIN Pitching ON HallOfFame.playerID = Pitching.playerID

    """
P_data = pd.read_sql(Pit_HOF, MLB)
P_data
```

Now we can perform our sum for each of the variables based on the name of the player.

```python
#We perform the sum for each player
Sum_pit = P_data.groupby('playerID').sum()
Sum_pit
```

### 3

It is necessary to create a table that matches the `HallOfFame` players with the `Batting` table and shows the values for `AB, R, H, HR, RBI, BB`, and `SO`.

```python
#Creating the query
Bat_HOF = """
    SELECT HallOfFame.playerID, AB, R, H, HR, RBI, BB, SO
    From HallOfFame
    Inner JOIN Batting ON HallOfFame.playerID = Batting.playerID
    """
B_data = pd.read_sql(Bat_HOF, MLB)
B_data
```

Now we compute the sum of the values by `playerID`.

```python
Sum_bat = B_data.groupby('playerID').sum()
Sum_bat
```

### 4

Now we are going to join the two tables by pitcher(`playerID`), using the `pd.merge` function.

```python
pit_join = pd.merge(Sum_pit, Sum_bat, on='playerID', how = 'inner')
pit_join
```
